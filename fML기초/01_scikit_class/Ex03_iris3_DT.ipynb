## 의사결정나무 (Decision Tree)


[예제 ] 붓꽃(Iris) 품종 

- 꽃잎(petal)과 꽃받침(sepal)의 폭과 길이를 측정하여 품종을 예측한다
- 150개의 데이타에서 3가지 품종(setosa, versicolor, virginica)로 분류한다

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# 1. 데이타 로딩

iris = datasets.load_iris()

# 데이터 key 확인
print(iris.keys())
```

```python
# 2. 데이터와 레이블 분리 변수 선언
X= iris.data # 데이터
y= iris.target # 레이블(답)

# 3. 데이타셋을 분리 ( 학습용:검증용 = 7:3 )

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.7)
```

```python
# 5. 트리 모델 생성하고 학습하기

from sklearn.tree import DecisionTreeClassifier
iris_tree= DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
iris_tree.fit(X_train,y_train)

```

```python
# 6. 정확도

print('학습용 데이터의 정확도 :{:.3f}'.format(iris_tree.score(X_train,y_train)))
print('학습용 데이터의 정확도 :{:.3f}'.format(iris_tree.score(X_test,y_test)))

```
