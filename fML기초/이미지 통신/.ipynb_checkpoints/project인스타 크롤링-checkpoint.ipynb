```python
import requests 
from selenium import webdriver as wd 
import time 
import re
from tqdm.notebook import tqdm

keyword = "서울맛집"

url = "https://www.instagram.com/explore/tags/{}/".format(keyword)

instagram_tags = []

driver = wd.Chrome("/KYE/Python/fML기초/이미지 통신/webdriver/chromedriver.exe")

# 검색페이지 접속
driver.get(url)
time.sleep(2)

# 첫번째 게시글 누르기
driver.find_element_by_css_selector('div.v1Nh3.kIKUG._bz0w').click()

# 로그인
id = '01097972948'
pw = 'qasxwedc1'

id_input = driver.find_elements_by_css_selector('#loginForm > div.Igw0E.IwRSH.eGOV_._4EzTm.kEKum > div:nth-child(1) > div > label > input')[0]
id_input.send_keys(id)
pw_input = driver.find_elements_by_css_selector('#loginForm > div.Igw0E.IwRSH.eGOV_._4EzTm.kEKum > div:nth-child(2) > div > label > input')[0]
pw_input.send_keys(pw)
pw_input.submit()

time.sleep(3)

# 로그인후 버튼
driver.find_element_by_class_name('sqdOP').click()

time.sleep(3)

#   첫번째 게시글 클릭
driver.find_element_by_css_selector("div._9AhH0").click()

time.sleep(0.6)

total_page = 5
for page in tqdm(range(1, total_page + 1)):
    time.sleep(0.6)

#   data = driver.find_element_by_class_name('C4VMK')
#   태그 불러오기
    data = driver.find_element_by_css_selector('body > div._2dDPU.CkGkG > div.zZYga > div > article > div.eo2As > div.EtaWk > ul > div > li > div > div > div.C4VMK') # C7I1f X7jCj
    
    tag_raw = data.text
    
    print(tag_raw)
    
    tags = re.findall('#[A-Za-z0-9가-힣]+', tag_raw)
    tag = ''.join(tags).replace("#"," ") # "#" 제거

    tag_data = tag.split()

    for tag_one in tag_data:
        instagram_tags.append(tag_one)
    print(instagram_tags)
    driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow').click()

    time.sleep(0.4)
driver.close()

```

```python
from collections import Counter 
count = Counter(instagram_tags) 
common_tag_200 = count.most_common(200)
```

```python
from PIL import Image

import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib

matplotlib.rcParams['font.family'] = "Maulgun Gothic"
font_path = "c:/Windows/Fonts/malgun.ttf"

# denne_mask = np.array(Image.open('cloud.png'))
custom_mask = np.array (Image.open ( "images.png")) 
wc = WordCloud(font_path=font_path,mask=custom_mask, background_color="white", width=800, height=600)
cloud = wc.generate_from_frequencies(dict(common_tag_200))
plt.figure(figsize = (20, 16))
plt.axis('off')
plt.imshow(cloud)
plt.savefig('./file/wordcloud.png')
```
