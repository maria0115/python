### 1. Series 의 plot()

```python
%matplotlib inline
```

```python
from pandas import Series
```

```python
# 판다스 그래프 : Series의 plot() 
s=Series([1.5, 2.3, 0.9], index=['no1','no2','no3'])

s.plot(kind='bar')
```

#### 위의 판다스 제공 그래프와 matplotlib 그래프 비교

```python
s=Series([1.5, 2.3, 0.9], index=['no1','no2','no3'])

splot = s.plot.bar()
splot.set_xlabel('data')
splot.set_ylabel('value')
splot.set_title('샘플')
```

## 2. DataFrame 그래프

```python
from pandas import DataFrame
import numpy as np

# 데이타프레임에서 시각화 작업
df=DataFrame(np.random.randn(10, 4),    # 난수 10개를 4번
             columns=['level','step','cnt','temp'], 
             index=np.arange(0,100,10))     # np.arange(0,100,10)
df.plot();
```

```python
df.plot(kind='barh',title='데이타 차트');
```
