```python
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc

plt.rcParams['axes.unicode_minus']=False  # 추가설정 : 폰트를 변경하면 -표시가 ㅁ으로 변경되기에 '-'를 변경하지 않도록 지정
rc('font', family='Malgun Gothic')
```

```python
# 기본적인 그래프
# data = [10,15,18,30,20]
# plt.figure(figsize=(12,6))
# plt.plot(data)
# plt.xlabel('시간')
# plt.ylabel('온도')
# plt.title('샘플')
# plt.show()

import pandas as pd
import matplotlib.pyplot as plt

dairy = pd.read_csv('data/dairy.csv')
data = dairy[dairy['year']>=2000]
data = data[data['dairy']=='치즈']
data
plt.figure(figsize=(12,6))
plt.plot(data['year'],data['production'])
plt.xlabel('year')
plt.title('치즈 판매량')
plt.grid(True)
plt.show()

data['year']
```

```python
# 1- 그래프 틀(영역)을 만들기
# 2- 그래프 그리기
# 3- 그래프 보여주기
plt.plot(data,color='blue',linestyle='dashed',marker='o')
```
