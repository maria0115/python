## 하나의 화면에 여러 개의 그래프 그려보기

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt



# plt.plot(np.random.randn(100))
# plt.plot(np.random.randn(100).cumsum())

fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax2.plot(np.random.randn(100))
ax1.plot(np.random.randn(100).cumsum())
```

```python
# x, y 축 값 범위 지정 

fig = plt.figure()
ax=fig.add_subplot(111)
ax.plot(np.random.randn(50),'b+--')
ax.set_xlim([1,100])
ax.set_ylim([-5,5])
ax.set_xlabel('data')
ax.set_ylabel('tem')
ax.set_title('my')
```

## 그래프 속성 지정

- title() : 
- xlabel() : 
- ylabel() : 
- gri() : 격자 눈금
- text() : 글씨를 출력 (tooltip을 간단하게 하고자 할 때)

```python
# plt에서 그래프 속성 지정
plt.plot(np.random.randn(10))
plt.xlabel('data')
plt.ylabel('tem')
plt.text(0.5,0.5,'Random Value')
plt.grid()
```

### x 축 간격

```python
# figure로 전체 그래프의 액자 같은 개념으로 액자 크기 조절이 가능하다
fig = plt.figure()
ax=fig.add_subplot(111)
ax.plot(np.random.randn(50),'b+--')
ax.set_xticks([0,10,20,30,40,50])
ax.set_xticklabels(['one','two','three','four','five','six'])
```
