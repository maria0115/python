# 이변량양적자료

> 1. 이변량 - 변수 2개

> 2. 양적자료 -   숫자형 자료         (ex. 온도, 키, 몸무게 )


            * 질적자료 = 명목형 = 이산형
            * 양적자료 = 수치형 = 연속형

> 3. 다변량 - 변수 여러개

```python
# 데이타 읽어오기
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# 인덱스가 2번 생성되기에 기존 인덱스를 인덱스로 지정
diamonds = pd.read_csv('data/diamonds.csv', index_col='Unnamed: 0')
```

```python
plt.scatter(diamonds['carat'],diamonds['price'])
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.scatter3D(diamonds['x'],diamonds['y'],diamonds['z'])
```
