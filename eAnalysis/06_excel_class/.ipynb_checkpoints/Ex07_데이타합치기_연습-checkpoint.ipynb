```python
import pandas as pd

transaction01 = pd.read_csv("data/transaction/201701.csv")

print(transaction01.shape)
transaction01.head()
```

```python
transaction02 = pd.read_csv("data/transaction/201702.csv")

print(transaction02.shape)
transaction02.head()
```

### 1. 두 데이타프레임을 합치기

- transaction01 과 transaction02 데이타 프레임 합치기
- concat() 이용

```python
# concat == concatenate
transaction = pd.concat([transaction01, transaction02])
transaction

# 나중에 지움
```

### 2. 두 데이타프레임 합치기

- transaction02와 cardstatement 데이타 프레임 합치기
- concat() 이용
- concat() 함수에 axis=0 이나 axis=1 지정

**concat() 명령어를 도움말 확인**

```python
import numpy as np
cardstatement = [
    {"payment_method": "카드 결제", "installment": "일시불"},
    {"payment_method": "카드 결제", "installment": "3개월"},
    {"payment_method": "무통장 입금", "installment": np.nan},
    {"payment_method": "카드 결제", "installment": "일시불"},
]

cardstatement = pd.DataFrame(cardstatement)
cardstatement
```

```python
transaction = pd.concat([transaction02, cardstatement], axis=1)
transaction

# 나중에 지움
```

### 3. 데이타합치기

- transaction과 user 데이타프레임을 합치기
- merge()


```python
user = pd.read_csv("data/transaction/user.csv")

print(user.shape)
user.head()
```

```python
# transaction, user 의 공통 컬럼은 "Name"
# "Name" 컬럼을 기준으로 병합(Merge) 한다.
# Default 는 Inner Merge 다. (교집합 )
pd.merge(transaction, user)


# 나중에 지우기
```

### merge() 연습

```python
# transaction01의 Yoon 데이터가 누락된다.
# user의 Name 컬럼에 Yoon이 없기 때문
pd.merge(transaction01, user)
```

```python
pd.merge(transaction01, user, how="inner") # 교집합
```

```python
pd.merge(transaction01, user, how="outer") # 합집합
```

```python
pd.merge(transaction01, user, how="left") # 좌측 기준
```

```python
pd.merge(transaction01, user, how="right") # 우측 기준
```
