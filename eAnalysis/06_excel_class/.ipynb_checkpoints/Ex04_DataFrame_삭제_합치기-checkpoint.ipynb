###  데이타프레임 합치기

1. append()

2. concat()

```python
df1 = pd.read_excel('data/2016-01.xls', index_col='상품명')
print(df1.shape)
print(df1.head())
df2 = pd.read_excel('data/2016-02.xls', index_col='상품명')
print(df1.shape)
print(df2.head())
```

```python
df1 + df2 # 각가의 인덱스별로 매칭만 하기에 의미 없음
```

```python
# 1. append()
# df1.append(df2) # 정렬관련 warning
df1.append(df2, sort=True) # warning 없어짐

df3 = df1.append(df2, sort=True)
print(df3.shape)
df3.head()
```

```python
# 2. concat()
# pd.concat([df1,df2], axis='index')
df4 =pd.concat([df1,df2], axis='index', sort=True)
print(df4.shape)
df4.head()
```
