## [ 연습 ]  datatime  오브젝트로 변경하여 필요한 데이타 추출 및 계산

```python
import pandas as pd


ebola = pd.read_csv('./data/ebola_timeseries.csv')
ebola.head()
ebola.info()
```

```python
# datetime 오브젝트로 변경하여 필요한 데이타 추출 및 계산

ebola['date_dt']=pd.to_datetime(ebola['Date'])
ebola.head()
```

```python
ebola['date_dt'].dt.year
```
