```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
# 데이타 로딩 - 기존 데이타셋과 다르기에 확인 필요
raw_data = pd.read_csv('./data/titanic.csv')
raw_data.info()

raw_data.describe()
```

##  생존자와 사망자의 시각화

```python
t = raw_data['Survived'].count()
d = raw_data['Survived'].count()-raw_data['Survived'].sum()
a = t-d

plt.pie([a,d])
d
```

# 연령분포 시각화

```python
e = raw_data.groupby('Age').count()
e
plt.scatter(e['Survived'],e.index)

```

# 객실등급별로 각 평균 통계 - 표

```python
t = raw_data.groupby('Pclass').mean()
t
plt.pie(t['Survived'],labels = t.index)
```

# 성별, 연령대, 객실등급별 생존자 시각화

```python
t = raw_data.groupby('Pclass').sum()
t
plt.bar(t.index,t['Survived'])

```

```python
t = raw_data.groupby('Pclass').sum()
t
plt.bar(t.index,t['Survived'])
```

```python
e = raw_data.groupby('Age').sum()
e
plt.scatter(e['Survived'],e.index)
```

```python
e = raw_data.groupby('Sex').sum()
e
plt.bar(e.index,e['Survived'])
```
