### 교통사고 데이타셋에서 요일별 사망자수 필터링 + 시각화

(1) 요일별 사망자수 필터링 (groupby() 이용 )

(2) 요일별 사망자수의 합계 (sum() 추가)

(3) 그래프

```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)
```

```python
# 데이타 읽어오기
data=pd.read_csv('data/accidentdata.csv')
data
```

```python
f = data.groupby('요일').sum()['사망자수']
# s = data.groupby('요일').count()['사망자수'].sum()
# plt.hist(f,labels=['월'])
label=['월','화','수','목','금','토','일']
index = np.arange(len(label))
f = f.reindex(['월','화','수','목','금','토','일'])
plt.bar(index,f)

plt.title('Sum of Day', fontsize=20)
plt.xticks(index, label, fontsize=15)

plt.show()
```
